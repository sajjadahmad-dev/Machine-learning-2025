```python
import requests
import pandas as pd

df_list = []

for i in range(1, 505):
    response = requests.get(
        f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}'
    )

    if response.status_code == 200:
        data = response.json()
        if 'results' in data:
            temp_df = pd.DataFrame(data['results'])[
                ['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']
            ]
            df_list.append(temp_df)
        else:
            print(f"⚠️ No 'results' key found on page {i}")
    else:
        print(f"❌ Failed to fetch page {i}: Status code {response.status_code}")

# Combine all valid pages
df = pd.concat(df_list, ignore_index=True)
print("✅ DataFrame created with shape:", df.shape)
```
